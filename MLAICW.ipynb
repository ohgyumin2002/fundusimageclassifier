```python
import numpy as np
import optuna
import seaborn as sns
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt

# Define paths
train_data_path = "eye-images1/dataset/train"
val_data_path = "eye-images1/dataset/val"
test_data_path = "eye-images1/dataset/test"

# Compute class weights (done once, reused in trials)
train_datagen = ImageDataGenerator(rescale=1.0/255.0)
train_generator_temp = train_datagen.flow_from_directory(
    train_data_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=True
)
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_generator_temp.classes),
    y=train_generator_temp.classes
)
class_weights = dict(enumerate(class_weights))

# Optuna objective function
def objective(trial):
    # Hyperparameters to tune
    learning_rate = trial.suggest_float('learning_rate', 1e-5, 1e-3, log=True)
    dropout_rate = trial.suggest_float('dropout_rate', 0.3, 0.5)
    dense_units = trial.suggest_int('dense_units', 64, 256, step=64)
    batch_size = trial.suggest_categorical('batch_size', [16, 32, 64])
    rotation_range = trial.suggest_float('rotation_range', 20, 50)
    zoom_range = trial.suggest_float('zoom_range', 0.2, 0.3)
    l2_weight = trial.suggest_float('l2_weight', 1e-6, 1e-4, log=True)

    # Data augmentation
    train_datagen = ImageDataGenerator(
        rescale=1.0 / 255.0,
        rotation_range=rotation_range,
        width_shift_range=0.2,
        height_shift_range=0.2,
        zoom_range=zoom_range,
        horizontal_flip=True,
        brightness_range=[0.9, 1.1],
        fill_mode='nearest'
    )
    val_datagen = ImageDataGenerator(rescale=1.0 / 255.0)

    train_generator = train_datagen.flow_from_directory(
        train_data_path,
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=True
    )
    val_generator = val_datagen.flow_from_directory(
        val_data_path,
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=False
    )

    # Simplified model definition
    base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    for layer in base_model.layers[:350]:  # Freeze more layers
        layer.trainable = False

    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(dense_units, activation='relu', kernel_regularizer=l2(l2_weight)),
        Dropout(dropout_rate),
        Dense(len(train_generator.class_indices), activation='softmax')
    ])

    # Compile the model
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    # Early stopping and learning rate scheduler
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)

    # Train the model
    history = model.fit(
        train_generator,
        validation_data=val_generator,
        epochs=30,  # Reduced epochs for faster optimization
        class_weight=class_weights,
        callbacks=[early_stopping, reduce_lr]
    )

    # Evaluate on validation data
    val_loss, val_accuracy = model.evaluate(val_generator, verbose=0)
    return val_loss  # Optuna minimizes validation loss

# Run Optuna optimization
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=30)  # Adjusted number of trials

# Print the best parameters
print("Best trial:")
print(study.best_trial.params)

# Train final model with best parameters
best_params = study.best_trial.params

# Final model training
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255.0,
    rotation_range=best_params['rotation_range'],
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=best_params['zoom_range'],
    horizontal_flip=True,
    brightness_range=[0.9, 1.1],
    fill_mode='nearest'
)
val_datagen = ImageDataGenerator(rescale=1.0 / 255.0)

train_generator = train_datagen.flow_from_directory(
    train_data_path,
    target_size=(224, 224),
    batch_size=best_params['batch_size'],
    class_mode='categorical',
    shuffle=True
)
val_generator = val_datagen.flow_from_directory(
    val_data_path,
    target_size=(224, 224),
    batch_size=best_params['batch_size'],
    class_mode='categorical',
    shuffle=False
)

# Define final model
base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
for layer in base_model.layers[:350]:  # Freeze more layers
    layer.trainable = False

final_model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(best_params['dense_units'], activation='relu', kernel_regularizer=l2(best_params['l2_weight'])),
    Dropout(best_params['dropout_rate']),
    Dense(len(train_generator.class_indices), activation='softmax')
])

# Compile final model
optimizer = Adam(learning_rate=best_params['learning_rate'])
final_model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train the final model
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)

final_model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=50,
    class_weight=class_weights,
    callbacks=[early_stopping, reduce_lr]
)

# Evaluate final model
test_datagen = ImageDataGenerator(rescale=1.0 / 255.0)
test_generator = test_datagen.flow_from_directory(
    test_data_path,
    target_size=(224, 224),
    batch_size=best_params['batch_size'],
    class_mode='categorical',
    shuffle=False
)

test_loss, test_accuracy = final_model.evaluate(test_generator)
print(f"Test Accuracy: {test_accuracy:.4f}")

# Predictions and metrics
y_true = test_generator.classes
y_pred = np.argmax(final_model.predict(test_generator), axis=-1)

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:")
print(cm)

# Classification report
class_labels = list(test_generator.class_indices.keys())
print("Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_labels, zero_division=1))

# Confusion Matrix Visualization
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle

# Binarize the output for multi-class ROC-AUC computation
y_true_bin = label_binarize(y_true, classes=list(range(len(class_labels))))
y_pred_prob = final_model.predict(test_generator)

# Compute ROC curve and ROC-AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(len(class_labels)):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC-AUC
fpr["micro"], tpr["micro"], _ = roc_curve(y_true_bin.ravel(), y_pred_prob.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Plot the ROC curves
plt.figure(figsize=(10, 8))
colors = cycle(['blue', 'red', 'green', 'purple', 'orange'])

for i, color in zip(range(len(class_labels)), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f"ROC curve for {class_labels[i]} (AUC = {roc_auc[i]:0.2f})")

# Plot micro-average ROC curve
plt.plot(fpr["micro"], tpr["micro"],
         label=f"micro-average ROC curve (AUC = {roc_auc['micro']:0.2f})",
         color='deeppink', linestyle=':', linewidth=4)

# Plot diagonal line (random classifier)
plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--")

# Customize the plot
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multi-class ROC-AUC Curve")
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()

from sklearn.metrics import f1_score

# Calculate weighted F1 score
weighted_f1 = f1_score(y_true, y_pred, average='weighted')

# Print the weighted F1 score
print(f"Weighted F1 Score: {weighted_f1:.4f}")
```
